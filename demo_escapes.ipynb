```python
'''
    sr -- sample rate
    audio -- исходный звук
    front, back -- элементы декомпозиции
    draw_plots -- рисует графики по желанию пользователя
    
'''
front, back, sr, audio = dc.decompose('speech_0088.wav',draw_plots=True)
```
